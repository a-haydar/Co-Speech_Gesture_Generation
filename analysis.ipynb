```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
import matplotlib.gridspec as gridspec

from grid import SeabornFig2Grid
```

```python
def display_joint_dist(df):
    fig = plt.figure(figsize=(17,12))
    gs = gridspec.GridSpec(3,3)

    cors = zip([df.columns[i] for i in range(0, len(df.columns), 2)], 
                [df.columns[i] for i in range(1, len(df.columns), 2)])
    for i, cor in enumerate(cors):
        x = cor[0]
        y = cor[1]
        g = sns.jointplot(x=x, y=y, data=df.loc[df.index % 20 == 0], 
                    kind="kde", space=0, color='g')
        SeabornFig2Grid(g, fig, gs[i])
    
    gs.tight_layout(fig)
    plt.show()
```

```python
def display_dist(df, drop=[3,4,7,6,15,16]):
    df = df.drop(drop, axis=1)
    for col in df.columns:
        g = sns.distplot(df.loc[df[col].notnull(), col],
                    label=str(col))
        # plt.title(col)
    plt.legend(title="Coordination", loc="upper left")
    return g
```

```python
data_origin = torch.load('./processed_data/l2_norm.pickle')
p_origin = pd.DataFrame(data_origin)
p_origin = p_origin.drop([i for i in range(2, len(p_origin.columns), 3)], axis=1)
p_origin.min()
```

```python
display_dist(p_origin, drop=[0,1,18,19,21,22,12,13,9,10,7,16,4])
```

```python
display_joint_dist(p_origin)
```
